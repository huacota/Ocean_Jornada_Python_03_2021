```python
#plt.figure(figsize=(15,5))
sns.countplot(data=df,x='education')
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='education')
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='education', palette='Reds')  
plt.ylabel('Contagem') #muda o nome do eixo vertical 
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='education', hue='marital')  
plt.ylabel('Contagem') #muda o nome do eixo vertical 
plt.show() #retirar  <matplotlib.axes._subplots.AxesSubplot at 0x7f2c80f32410>

#obs: veja q % de universitarios é muito alta e de analfabeto baixissima, 
#isso pode ser um viez pq pode ser q pesssoas analfabetaas podem n ter acesso a telefone 
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='age', hue='marital')  #melhor para dados categoricos, pega todas as idades possíveis
plt.ylabel('Contagem') #muda o nome do eixo vertical 
plt.show() #retirar  <matplotlib.axes._subplots.AxesSubplot at 0x7f2c80f32410>
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='job', hue='education')  #melhor para dados categoricos 
plt.ylabel('Contagem') #muda o nome do eixo vertical 
plt.show() #retirar  <matplotlib.axes._subplots.AxesSubplot at 0x7f2c80f32410>
```
